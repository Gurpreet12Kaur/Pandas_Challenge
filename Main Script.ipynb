```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
##setting all variables
##Total_schools
total_schools = len(school_data_complete['school_name'].unique()) 
##total_students
total_students = len(school_data_complete['Student ID'].unique())
##print(total_students)
##print(total_schools)
##total_budget
total_budget = school_data_complete['budget'].unique()
total_budget_sum = total_budget.sum()
##print(total_budget_sum)
##maths average
average_math = round(school_data_complete['maths_score'].mean(),6)
##print(average_math)
##read average
average_reading = round(school_data_complete['reading_score'].mean(),6)
##print(average_reading)
#% maths
math_score = school_data_complete['maths_score'].tolist()
passing_score = 0
total_students=len(math_score)

for score in math_score:
    if score >=50:
        passing_score = passing_score+1
passing_math_percentgae = round((passing_score/total_students)*100,6)
##print(passing_math_percentgae)

#reading score
read_score = school_data_complete['reading_score'].tolist()
passing_score = 0
total_students=len(read_score)

for score in read_score:
    if score >=50:
        passing_score = passing_score+1
passing_reading_percentgae = round((passing_score/total_students)*100,6)

##print(passing_reading_percentgae)

#overall percentage
passing_math = school_data_complete['maths_score'] >= 50  # Example passing criteria for math
passing_reading = school_data_complete['reading_score'] >= 50  # Example passing criteria for reading


overall_passing = passing_math & passing_reading  
overall_passing_percentage = round((overall_passing.sum() / len(school_data_complete)) * 100,6)

##print(overall_passing_percentage)



```

```python
##LOCAL GOVERNMENT AREA SUMMARY
area_summary=pd.DataFrame({
    "Total Schools":[total_schools],
    "Total Student":[total_students],
     "Total Budget": ["${:,.2f}".format(total_budget_sum)],
    "Average Math Score":[average_math],
    "Average Reading Score":[average_reading],
    "% Passing Maths":[passing_math_percentgae],
    "% Passing Reading":[passing_reading_percentgae],
    "% Overall Passing":[overall_passing_percentage]
})


area_summary
```

```python
##setting all variables
##school_count
school_count= len(school_data_complete["school_name"].unique())

#School Name
school_types=school_data.set_index(["school_name"])["type"]
school_types.head()

#Total Students in each school
student_per_school=school_data_complete["school_name"].value_counts()
student_per_school

#Budget
school_budget=school_data_complete.groupby(["school_name"])["budget"].mean()
school_budget.head()

#Per Student Budget
per_student_budget=school_budget/student_per_school
per_student_budget.head()

#Average Math Score per school
average_math_perSchool=school_data_complete.groupby(["school_name"])["maths_score"].mean()
average_math_perSchool.head()

#Average Reading Score per school
average_reading_perSchool= school_data_complete.groupby(["school_name"])["reading_score"].mean()
average_reading_perSchool.head()



#% Passing Math per school for % passing math
passing_math_score_perSchool=school_data_complete[school_data_complete["maths_score"]>=50]
passing_math_score_perSchool

#% passing math
group_math_school=passing_math_score_perSchool["school_name"].value_counts() 
group_math_school.head()
percent_math=(group_math_school/student_per_school)*100
percent_math

# % Passing Reading per school for % passing reading
read_scor=school_data_complete[school_data_complete["reading_score"]>=50]
by_read_PerSchool= read_scor["school_name"].value_counts()
by_read_PerSchool.head()
percent_read=(by_read_PerSchool/student_per_school)*100
percent_read

# % Overall Passing Rate
passing_math = school_data_complete['maths_score'] >= 50  
passing_reading = school_data_complete['reading_score'] >= 50  
passing_mathsandreading = school_data_complete[(school_data_complete['maths_score'] >= 50)  & (school_data_complete['reading_score'] >= 50)]
overall_Score=passing_mathsandreading.groupby(["school_name"]).size()
overall=overall_Score/student_per_school*100
```

```python
per_school_summary_df = pd.DataFrame({"School Type":school_types, 
                                 "Total Students":student_per_school, 
                                 "Total School Budget":school_budget,
                                 "Per Student Budget":per_student_budget,
                                 "Average Math Score" :average_math_perSchool,
                                 "Average Reading Score":average_reading_perSchool,
                                 "% Passing Math":percent_math, 
                                 "% Passing Reading":percent_read,
                                 "% Overall Passing Rate":overall
                                 })

per_school_summary_df["Total School Budget"] = per_school_summary_df["Total School Budget"].map("${:,.2f}".format)
per_school_summary_df.head(15)
```

```python
# Convert the 'year' column to numeric data type 
school_data_complete['year'] = pd.to_numeric(school_data_complete['year'])

# Initialize a dictionary to store the average math scores for each year
yearly_math_scores = {}

# Loop through unique years
for year in school_data_complete['year'].unique():
  
    year_data = school_data_complete[school_data_complete['year'] == year]
    
    # Calculate the mean math score for each school in the current year
    year_math_scores = year_data.groupby('school_name')['maths_score'].mean()
    
    # Store the average math scores for the current year in the dictionary
    yearly_math_scores[year] = year_math_scores


math_scores_per_year = pd.DataFrame(yearly_math_scores)

math_scores_per_year = math_scores_per_year[[
                            9, 10, 11, 12  
                                            ]]

math_scores_per_year.columns = ["Year 9", "Year 10", "Year 11", "Year 12"]


math_scores_per_year.index.name = ' '

# Display the DataFrame
math_scores_per_year.head(15)

#yr9 -
```

```python
school_data_complete['year'] = pd.to_numeric(school_data_complete['year'])

yearly_reading_scores = {}

# Loop through unique years
for year in school_data_complete['year'].unique():

    year_data = school_data_complete[school_data_complete['year'] == year]
    
    # Calculate the mean reading score for each school in the current year
    year_read_scores = year_data.groupby('school_name')['reading_score'].mean()
    
    # Store the average reading scores for the current year in the dictionary
    yearly_reading_scores[year] = year_read_scores

# Create a DataFrame from the dictionary
reading_scores_per_year = pd.DataFrame(yearly_reading_scores)

reading_scores_per_year = reading_scores_per_year[[
                                         9, 10, 11, 12  
                                                 ]]


reading_scores_per_year.columns = ["Year 9", "Year 10", "Year 11", "Year 12"]

reading_scores_per_year.index.name = ' '

reading_scores_per_year.head(15)
```
